```python
my_songs = pd.read_pickle("./my_songs.pkl")
public_df = pd.read_pickle("./public_df.pkl")
public_df.head()
```

```python
# get genre counts for public songs
public_genres = public_df.loc[:,['genre', 'genre count']].drop_duplicates().reset_index(drop=True)
print('\n public song genre counts:')
display(public_genres.head())

# get genre counts for my songs (score goes negative for dislikes)
my_genres = my_songs.loc[:,['genre', 'genre count','Liked']].reset_index(drop=True).drop_duplicates()
my_genres["Liked"].replace({0: -1}, inplace=True)
my_genres['my genre score'] = my_genres['genre count']*my_genres['Liked']
del my_genres['genre count']
del my_genres['Liked']
my_genres = my_genres.groupby('genre').sum()

# combine genres
combined_genres = public_genres.merge(my_genres, on = 'genre', how = 'inner')
del combined_genres['genre count']
print('\n my score for public song genres (count * like):')
display(combined_genres.head())
public_df_2 = public_df.reset_index().merge(combined_genres, on = 'genre', how = 'inner').set_index('id')

print('\n public df with my songs associated score:')
display(public_df_2.head())

print('\n my songs df:')
display(my_songs.head())
# public_df[['release_year']] = public_df[['release_year']].apply(pd.to_numeric) 
# print(public_df.dtypes) 
```

```python
# still need to adjust public df's genre agg score
genre_agg_score = pd.DataFrame(public_df_2.loc[:,'my genre score']).groupby('id').sum()
genre_agg_score.columns = ['genre agg score']
del public_df_2['genre agg score']
public_df_3 = public_df_2.reset_index().merge(genre_agg_score, on = 'id', how = 'left').set_index('id')
display(public_df_3.head())
display(public_df_3.loc['4GiXBCUF7H6YfNQsnBRIzl']) # to make sure it's working properly
```

```python
# set up data for training
# x values = one hot encoded genres
setup = my_songs.loc[:,['genre']]
print(setup.shape)
display(setup.head())
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder

# one hot encode features such as artist
x_dum = pd.get_dummies(setup).reset_index()
x_dum = x_dum.groupby('id').sum()
print(x_dum.shape)
display(x_dum.head())
# and genre agg score (sum of all genres for 1 song) and liked indicator
my_genre_agg_score = my_songs.loc[:,['genre agg score','Liked']].reset_index().drop_duplicates().set_index('id')
x_dum = x_dum.merge(my_genre_agg_score, on = 'id', how = 'inner')
print(x_dum.shape)
display(x_dum.head())
y=x_dum['Liked']
```

```python
# set up data for validation
# x values = one hot encoded genres
setup_val = public_df_3.loc[:,['genre']]
print(setup_val.shape)
display(setup_val.head())

# one hot encode features such as artist
x_dum_val = pd.get_dummies(setup_val).reset_index()
x_dum_val = x_dum_val.groupby('id').sum()
print(x_dum_val.shape)
display(x_dum_val.head())
# and genre agg score (sum of all genres for 1 song) and liked indicator
public_genre_agg_score = public_df_3.loc[:,['genre agg score']].reset_index().drop_duplicates().set_index('id')
x_dum_val = x_dum_val.merge(public_genre_agg_score, on = 'id', how = 'inner')
print(x_dum_val.shape)
display(x_dum_val.head())
```

```python
# make feature columns a set of feature columns from my songs and public songs (set is the intersection of both)
feat_cols_new = list(set(x_dum.columns) & set(x_dum_val.columns))
print(len(x_dum.columns))
print(len(x_dum_val.columns))
print(len(feat_cols_new)) # combined genres in previous step
print(feat_cols_new[:30])
```

```python
# scale features
scaler = MinMaxScaler()
X = scaler.fit_transform(x_dum[feat_cols_new])
```

```python
X.shape
```

```python
X, y = resample(X, y, random_state=0)

# split into train, test groups
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
# define model parameters
pbounds = {
    'learning_rate': (0.01, 1.0),
    'n_estimators': (100,1000),
    'max_depth': (3,10),
    'subsample': (1.0, 1.0),  # Change for big datasets
    'colsample': (1.0, 1.0),  # Change for datasets with lots of features
    'gamma': (0, 5)}
def xgboost_hyper_param(learning_rate,
                        n_estimators,
                        max_depth,
                        subsample,
                        colsample,
                        gamma):
    max_depth = int(max_depth)
    n_estimators = int(n_estimators)
    clf = XGBClassifier(
        max_depth=max_depth,
        learning_rate=learning_rate,
        n_estimators=n_estimators,
        gamma=gamma)
    return np.mean(cross_val_score(clf, X_train, y_train.values.ravel(), cv=3, scoring='roc_auc'))

optimizer = BayesianOptimization(
    f=xgboost_hyper_param,
    pbounds=pbounds,
    random_state=1,
)

optimizer.maximize(
    init_points=3,
    n_iter=2,
)

# for i, res in enumerate(optimizer.res):
#     print("Iteration {}: \n\t{}".format(i, res))

print(optimizer.max)
```

```python
# define model and tune according to optimal parameters discovered above
model = xgb.XGBClassifier(
    #objective="binary:logistic", 
    random_state=42,
    colsample=int(optimizer.max['params']['colsample']),
    gamma=int(optimizer.max['params']['gamma']),
    learning_rate=optimizer.max['params']['learning_rate'],
    max_depth=int(optimizer.max['params']['max_depth']),
    n_estimators=int(round(optimizer.max['params']['n_estimators'],0)),
    subsample=int(optimizer.max['params']['subsample'])
    ).fit(X_train, y_train.values.ravel())
```

```python
# see how model does on test data
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred, labels=[0,1]))
```

```python
# use model to predict liked songs on actual data
print(x_dum_val.shape)
X_pred = scaler.fit_transform(x_dum_val[feat_cols_new])
predictions = model.predict_proba(X_pred)
```

```python
x_dum_val['Prediction'] = predictions[:,0]
x_dum_val['Predicted_Like'] = predictions[:,0] >= 0.99
pred_likes = pd.DataFrame(x_dum_val.loc[:,'Predicted_Like']).merge(public_df_3, on = 'id')
pred_likes_2 = pred_likes.loc[:,['Predicted_Like', 'album', 'artist_uri', 'artist', 'track', 'url',
       'popularity', 'release_date', 'release_year',
       'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
       'duration_ms', 'time_signature', 'genre agg score']].drop_duplicates()
# genre_df = pd.DataFrame(np.unique(pred_likes.loc[:,'genre list']))
# genre_df.columns = ['genre list']
# display(genre_df.head())
display(pred_likes_2.head())
pred_likes_3 = pred_likes_2[pred_likes_2['genre agg score'] > 131]
print('pred_likes_3:')
display(pred_likes_3.head())
pred_likes_4 = pred_likes_3
#pred_likes_4 = pred_likes_3[pred_likes_3['Predicted_Like'] == 1]
display(pred_likes_4.head())
print(pred_likes_4.shape)
```
